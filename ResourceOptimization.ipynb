```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter


```

```python
x = np.arange(0,100,5)
k = np.arange(1,21,1, dtype=float)

X,K = np.meshgrid(x,k)

def z_func(X,K):
    return X**(-K)

z = z_func(X,K)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, k, z)
plt.xlabel('x')
plt.ylabel('k')
plt.show()
```

```python
a = np.arange(1,30,1)
b = 3

def c_func(a,b):
    return 1/(a**b)
c = c_func

plt.plot(a,c(a,b))
plt.xlabel('a')
plt.ylabel('c')
```
